```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=150,n_features=2,centers=3,cluster_std=0.5,shuffle=True,random_state=0)
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],c='white',edgecolor='black',marker='o',s=50)
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,init='random',max_iter=100,n_init=10)
y_km=km.fit_predict(x)
```

```python
print('distortion=',km.inertia_)
```

```python
print(km.cluster_centers_)
```

```python
plt.figure()
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='lightgreen',marker='d',edgecolor='black',label='cluster-1')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='orange',marker='o',edgecolor='black',label='cluster-2')
plt.scatter(x[y_km==2,0],x[y_km==2,1],s=50,c='yellow',marker='>',edgecolor='black',label='cluster-3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='red',marker='*',edgecolor='black',label='centroid')
plt.legend()
plt.show()
```

```python
dist=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=100)
  km.fit(x)
  dist.append(km.inertia_)
plt.plot(range(1,11),dist,marker='o')
plt.xlabel('numer of clusters (K)-->')
plt.ylabel('distortioans')
plt.show()
```

```python
import numpy as np
x=np.array([[1,2],[1,4],[1,0],[4,2],[4,4],[4,0]])
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
z=linkage(x,'single')
dendrogram(z)
plt.title('hierarchial clustring')
plt.xlabel('data points')
plt.ylabel('distance')
plt.show
```

```python
from sklearn.datasets import make_moons
x,y=make_moons(n_samples=200,noise=0.05,random_state=0)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,random_state=0)
y_km=km.fit_predict(x)
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='lightgreen',marker='o')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='blue',marker='d')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=50,c='red',marker='*')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
km=AgglomerativeClustering(n_clusters=2,linkage='single')
y_km=km.fit_predict(x)
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='lightgreen',marker='o')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='blue',marker='d')

plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
km=DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_km=km.fit_predict(x)
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='lightgreen',marker='o')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='blue',marker='d')

plt.legend()
plt.show()
```
